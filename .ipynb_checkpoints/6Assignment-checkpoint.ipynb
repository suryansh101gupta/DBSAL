```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
ds = sns.load_dataset('iris')
```

```python
fig, axes=plt.subplots(2,2,figsize=(15,9))
sns.histplot(ds['sepal_length'], ax=axes[0,0])
sns.histplot(ds['sepal_width'], ax=axes[0,1])
sns.histplot(ds['petal_length'], ax=axes[1,0])
sns.histplot(ds['petal_width'], ax=axes[1,1])
```

```python
fig, axes=plt.subplots(2,2,figsize=(15,9))
sns.boxplot(y='sepal_length', x='species', data=ds, ax=axes[0,0], hue='species')
sns.boxplot(y='sepal_width', x='species', data=ds, ax=axes[0,1], hue='species')
sns.boxplot(y='petal_length', x='species', data=ds, ax=axes[1,0], hue='species')
sns.boxplot(y='petal_width', x='species', data=ds, ax=axes[1,1], hue='species')
```
