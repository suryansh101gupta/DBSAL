```python
from scipy import stats
```

```python
z = np.abs(stats.zscore(ds['math score']))
print(z)
```

```python
threshold = 1
sample_outliers = np.where(z >threshold)
print(sample_outliers)
# threshold2 = 0.2
# sample_outliers2 = np.where(z <threshold2)
# print(sample_outliers2)
```

```python
sorted_mscore = sorted(ds['math score'])
sorted_mscore
```

```python
q1 = np.percentile(sorted_mscore, 25)
q3 = np.percentile(sorted_mscore, 75)
print(q1,q3)
```

```python
IQR = q3-q1
print(IQR)
```

```python
lwr_bound = q1-(1.5*IQR)
upr_bound = q3+(1.5*IQR)
print(lwr_bound, upr_bound)
```

```python
m_outliers = []
for i in sorted_mscore:
    if (i<lwr_bound or i>upr_bound):
        m_outliers.append(i)
print(m_outliers)
```

```python
#TRIMMING THE OUTLIER
# new_ds=ds.copy()
# for i in sample_outliers:
#     new_ds.drop(i,inplace=True)
# new_ds
```

```python
# QUANTILE BASED FLOORING AND CAPPING
ds_stud=ds.copy()
ninetieth_percentile = np.percentile(ds_stud['math score'], 90)
b = np.where(ds_stud['math score']>ninetieth_percentile, ninetieth_percentile, ds_stud['math score'])
print("New array:",b)
```

```python
ds_stud.insert(1,"m score",b,True)
ds_stud
```

```python
# Mean/Median imputation:
col = ['math score']
ds.boxplot(column = col)
plt.show()
```

```python
median=np.median(sorted_mscore)
median
```

```python
refined_ds=ds.copy()
refined_ds['math score'] = np.where(refined_ds['math score'] > upr_bound, median,refined_ds['math score'])
refined_ds['math score'] = np.where(refined_ds['math score'] < lwr_bound, median,refined_ds['math score'])
```

```python
refined_ds
```

```python
col = ['math score']
refined_ds.boxplot(column = col)
plt.show()
```
