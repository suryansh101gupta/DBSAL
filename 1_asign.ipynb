```python
iris['Species'].unique()
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
iris['Species'] = label_encoder.fit_transform(iris['Species'])
```

```python
iris['Species'].unique()
```

```python
label_encoder.classes_
```

```python
iris['Species'] = label_encoder.inverse_transform(iris['Species'])
```

```python
features_df=iris.drop(columns=['Species'])
enc = preprocessing.OneHotEncoder()
enc_df=pd.DataFrame(enc.fit_transform(iris[['Species']]).toarray())
df_encode = features_df.join(enc_df)
df_encode.rename(columns = {0:'Iris-Setosa', 1:'Iris-Versicolor',2:'Iris-virginica'}, inplace = True)
df_encode
```
